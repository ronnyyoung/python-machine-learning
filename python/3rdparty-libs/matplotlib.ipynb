# Matplotlib 画图

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 线状图

```python
x = np.linspace(0,2, 5)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# 设置坐标轴代表的意思
plt.xlabel('x label')
plt.ylabel('y label')

# 设置坐标轴的最小值与最大值
#plt.axis([-1,3, -2, 10])

# 设置标题
plt.title('Simple Plot')

# 设置不同曲线代表的意义
plt.legend()

plt.show()
```

## 自定制线条格式

```python
x = np.linspace(0, 2, 20)
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g^')
plt.show()
```

## 多图

```python
t1 = np.arange(0, 5.0, 0.1)
t2 = np.arange(0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, np.exp(-t1) * np.cos(2 * np.pi * t1), 'bo')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'k')

plt.show()
```

## 柱状图

```python
x1 = [1, 3, 5, 7]
y1 = [12, 6, 31, 10]
x2 = [i + 0.5 for i in x1]
y2 = [3, 7, 21, 16]
plt.bar(x1, y1, width=0.5, align='center', color='coral')
plt.bar(x2, y2, width=0.5)
plt.show()
```

## 直方图统计

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha = 0.75)

plt.xlabel('Smarts', fontsize=14, color='red')
plt.ylabel('Probability')
plt.title('Hisgogram of IQ')

plt.text(60, 0.025, r'$\mu=100, \ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```
